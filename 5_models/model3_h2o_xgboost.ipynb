```python
import h2o
from h2o.estimators import H2OXGBoostEstimator
h2o.init()
```

```python
df = h2o.import_file("/content/drive/MyDrive/Colab Notebooks/HACKATHON NOV22/MODEL3 - H2O train_test_split+balance+validation/train_labelencoded_standardized_balanced.csv")
df.describe()
```

```python
df['final_status'] = df['final_status'].asfactor()
```

```python
predictors = df.columns
predictors.remove('C1')
predictors.remove('final_status')
response = "final_status"
```

```python
train, valid = df.split_frame(ratios=[.8], seed=1234)
```

```python
df_xgb = H2OXGBoostEstimator(booster='dart',
                                  normalize_type="tree",
                                  seed=1)
```

```python
df_xgb.train(x=predictors,
                  y=response,
                  training_frame=train,
                  validation_frame=valid)
```

```python
df_xgb.F1()
```

```python
import pandas as pd
f1 = pd.DataFrame(df_xgb.F1(), columns=['threshold', 'f1'])
f1
```

```python
f1.to_csv('/content/drive/MyDrive/Colab Notebooks/HACKATHON NOV22/MODEL3 - H2O train_test_split+balance+validation/metrics_xgboost.csv')
```
